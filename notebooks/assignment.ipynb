```python
names = np.array(["Ana", "Ben", "Chen", "Dana", "Eli", "Fatima", "George", "Hui"])
spend = np.array([200, 150, 300, 120, 180, 220, 160, 310])      # marketing spend
revenue = np.array([400, 180, 500, 100, 220, 260, 150, 600])   # revenue

#Compute the ROI for each person: roi = revenue / spend
roi = revenue / spend
print("ROI for each person:", roi)  
```

```python
roi_mask = roi >= 2.0
high_roi_names = names[roi_mask]
print("People with ROI >= 2.0:", high_roi_names)

#List their spend and revenue as well
high_roi_spend = spend[roi_mask]
high_roi_revenue = revenue[roi_mask]
print("Spend for people with ROI >= 2.0:", high_roi_spend)
print("Revenue for people with ROI >= 2.0:", high_roi_revenue)
```

```python
spend_mask = spend >= 200
high_spend_names = names[spend_mask]
print("People with spend >= 200:", high_spend_names)

roi_and_spend_mask = (roi >= 2.0) & (spend >= 200)
high_roi_and_spend_names = names[roi_and_spend_mask]
print("People with ROI >= 2.0 and spend >= 200:", high_roi_and_spend_names)
```

```python
students = np.array(["S1", "S2", "S3", "S4", "S5"])
subjects = np.array(["Math", "Stats", "Python"])

scores = np.array([
    [75, 80, 85],
    [60, 65, 70],
    [90, 88, 92],
    [82, 79, 84],
    [70, 72, 78]
])

scores_avg_student = scores.mean(axis=1)
print("Average score for each student:", scores_avg_student)
scores_avg_subject = scores.mean(axis=0)
print("Average score for each subject:", scores_avg_subject)
scores_centered = scores - scores_avg_subject
print("Centered scores:\n", scores_centered)

#compare which student looks best by raw average vs centered average
print("Raw Averages:", scores_avg_student)
print("Centered Averages:", scores_centered.mean(axis=1))
```

```python
daily_visits = np.array([
    120, 130, 125, 140, 150,
    160, 170, 155, 145, 135,
    128, 132, 138, 142, 148,
    152, 158, 162, 168, 172,
    180, 190, 185, 175, 165,
    155, 145, 135, 125, 115
])

weekly_visits = daily_visits.reshape(6, 5) # Reshape to 6 weeks (rows) and 5 days (columns)
print("Weekly visits:\n", weekly_visits)

weekly_visits = weekly_visits.sum(axis=1)  # Sum across columns to get total visits per week
print("Total visits per week:", weekly_visits)
```

```python
avg_day_of_week = weekly_visits.mean(axis=0)
print("Average visits per day of the week:", avg_day_of_week)
```

```python
# Flatten
original_shape = weekly_visits.flatten()
print("Back to original match?", np.array_equal(daily_visits, original_shape))
```
